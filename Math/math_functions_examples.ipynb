```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

1. Linear Functions (x * k)

a function maps values of x from set of numbers d to values of y from set of values big y

```python
x = 1
y = 2 * x
print(f'when x = {x}, then y = {y}')

x = 5
y = 5 * x
print(f'when x = {x}, then y = {y}')

#plotting the graph of y = 2 * x

x_values = np.linspace(0,10,100)
y_values = 2 * x_values

plt.plot(x_values, y_values, label='y = 2 * x')
plt.title('The graph of y = 2x')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid(True)
plt.show()
```

```python
#plotting multiple values of k
k = [-1,1,2,3,4]
x_values = np.linspace(-10,10,100)

plt.figure()
for k in k:
    y_values = k * x_values
    plt.plot(x_values, y_values, label=f'y={k}*x')

plt.title('The graphs of y = k * x')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid(True)
plt.show()
```

2. Quadratic Functions

quadratic are non-linear functions that map all numbers postive or negative x to some positive y value

quadratic accelerate the rate of growth of the mapped y as x grows

```python
nums = [-3,-1,0,1,3]
for num in nums:
    y = num **2
    print(f'when x = {num}, y^2 = {y}')

x_values = np.linspace(-10,10,100)
y_values = x_values ** 2

plt.figure()
plt.plot(x_values, y_values, label='y= x^2', color='red')
plt.xlabel('x')
plt.axhline(y=0, color='k')
plt.ylabel('y')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title('the graph of y = x^2')
plt.show()
```

3. Cubic Functions

cubic function apply the cube function to values of x

As with the difference between linear and quadratic, values of y grow at accelerated rate with the same x when comparing 
cubic functions and quadratic functions

```python
nums = [-3,-1,0,1,3]
for num in nums:
    y = num **3
    print(f'when x = {num}, y^3 = {y}')

x_values = np.linspace(-10,10,100)
y_values = x_values ** 3

plt.figure()
plt.plot(x_values, y_values, label='y= x^3', color='green')
plt.xlabel('x')
plt.axhline(y=0, color='k')
plt.ylabel('y')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title('the graph of y = x^3')
plt.show()
```

4. Square Root Function

instead of having a positive integer exponent, the square root function map x to y by square rooting values of x.
in other terms the exponent is 1/2.

the square root function increase rapidly for small values of x and slows down as x grows larger

```python
nums = [0,1,4,9,16,25]
for num in nums:
    y = np.sqrt(num)
    print(f'when x = {num}, y = {y}')

x_values = np.linspace(0,25,100)
y_values = np.sqrt(x_values)

plt.figure()
plt.plot(x_values, y_values, label='y= sqrt(x)', color='purple')
plt.xlabel('x')
plt.axhline(y=0, color='k')
plt.ylabel('y')
plt.axvline(x=0, color='k')
plt.grid(True)
plt.title('the graph of y = sqrt(x)')
plt.show()
```

5. Exponential functions

exponential function have a x value raised to some integer value k. in this case we will be using 2 as our k value.

exponential functions increase faster than quadratic functions

```python
nums = [-1,0,1,3 ,5,7]
for num in nums:
    y = 2 **num
    print(f'when x = {num}, y = {y}')

x_values = np.linspace(-10,10,100)
y_values = 2 ** x_values

plt.figure()
plt.plot(x_values, y_values, label='y= 2^x_values', color='green')
plt.xlabel('x')
plt.axhline(y=0, color='k', linewidth=0.5)
plt.ylabel('y')
plt.axvline(x=0, color='k', linewidth=0.5)
plt.grid(True)
plt.title('the graph of y = 2^x_values')
plt.show()
```

6. Trigometric Functions

```python
import math

nums = [0, math.pi/2, math.pi, 3*math.pi/2, 2*math.pi]
for num in nums:
    y =math.sin(num)
    print(f'when x = {num:.2f}, y = {y:.2f}')

x_values = np.linspace(0,4*np.pi,100)
y_values = np.sin(x_values)

plt.figure()
plt.plot(x_values, y_values, label='y= sin(x)', color='red')
plt.xlabel('x')
plt.axhline(y=0, color='k', linewidth=0.5)
plt.ylabel('y')
plt.axvline(x=0, color='k', linewidth=0.5)
plt.grid(True)
plt.title('the graph of y = sin(x)')
plt.show()
```

7. Linear functions (y = m*x + b)

linear functions work the same as seen previously however there is now a constant value b. The effect that b has on the function is to move up or down the value of y by b.

```python
nums = [-1,0,1,3 ,5,7]


x_val = np.linspace(-10,10,100)
nums = [-1,0,1,3 ,5,7]
for num in nums:
    y = 2 *x_val + num
    plt.plot(x_val, y, label=f'2*x_values + {num}')

plt.xlabel('x')
plt.axhline(y=0, color='k', linewidth=0.5)
plt.ylabel('y')
plt.axvline(x=0, color='k', linewidth=0.5)
plt.grid(True)
plt.title('the graph of y = 2*x_values + num')
plt.show()
```
