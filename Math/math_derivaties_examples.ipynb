2. Derivatives

The derivatives of a function tells how y is affected by small inceases in x, in other words it tells us the rate at which the function increases when there is a change in x.

```python
#The numerical formula to work out the derivatie

def numerical_derivative(f, x, h=0.0001):
    ยง    return (f(x + h) - f(x)) / (2*h)
```

```python
def f(x):
    return 2 * x

x = np.linspace(1,10,100)
y = f(x)
#this is read as the derivate of function y with respect to x
dy_dx = np.full_like(x, 2) 

#np.full_like creates an array with the same shape as the passed in array and fills it with the value of the second argument


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,5))

#Function Plot 
ax1.plot(x, y, 'b-', linewidth = 2, label='y = 2x')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('The function y = 2x')
ax1.grid(True)
ax1.legend()

#Derivative plot
ax2.plot(x, dy_dx, 'r-',linewidth = 2, label='dy/dx = 2')
ax2.set_xlabel('x')
ax2.set_ylabel('dy/dx')
ax2.set_title('The derivative dy/dx = 2')
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()
```

The derivativ of y = x ^ 2

with quadratic functions the gradient changes as the x value changes, meaning at some points on a quadratic curve the gradient is slower compared to other points

```python
def f(x):
    return x ** 2

def dy_dx(x):
    return 2 * x

x = np.linspace(-10,10,100)
y = f(x)
#this is read as the derivate of function y with respect to x
dy_dx = dy_dx(x)

#np.full_like creates an array with the same shape as the passed in array and fills it with the value of the second argument


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,5))

#Function Plot 
ax1.plot(x, y, 'b-', linewidth = 2, label='y = x^2')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('Function y = x^2')
ax1.grid(True)
ax1.legend()

#Derivative plot
ax2.plot(x, dy_dx, 'r-',linewidth = 2, label='dy/dx = 2x')
ax2.set_xlabel('x')
ax2.set_ylabel('dy/dx')
ax2.set_title('Derivative dy/dx = 2x')
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
# Draw tangent lines at different points on y = xยฒ
def f(x):
    return x ** 2

def df_dx(x):
    return 2 * x

def tangent_line(x, x0):
    """Equation of tangent line at x0: y = f(x0) + f'(x0)(x - x0)"""
    return f(x0) + df_dx(x0) * (x - x0)

x = np.linspace(-3, 3, 100)
y = f(x)

# Points where we'll draw tangent lines
tangent_points = [-2, -1, 0, 1, 2]

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2, label='y = xยฒ')

# Draw tangent lines
for x0 in tangent_points:
    x_tangent = np.linspace(x0 - 1, x0 + 1, 50)
    y_tangent = tangent_line(x_tangent, x0)
    slope = df_dx(x0)
    plt.plot(x_tangent, y_tangent, 'r--', linewidth=1.5, alpha=0.7)
    plt.plot(x0, f(x0), 'ro', markersize=8)
    plt.text(x0 + 0.2, f(x0) + 0.5, f'slope={slope}', fontsize=9)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Tangent Lines on y = xยฒ')
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```

The derivative of y = mx + b

when finding the derivative of a function y with respect to x that contains a constant b, the b term will vanish.
This is because b isnt a coefficent of x thus a small change in x wont affect b influence on y.

```python
def f(x):
    return 2 * x + 3

x = np.linspace(1,10,100)
y = f(x)
#this is read as the derivate of function y with respect to x
dy_dx = np.full_like(x, 2) 

#np.full_like creates an array with the same shape as the passed in array and fills it with the value of the second argument


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,5))

#Function Plot 
ax1.plot(x, y, 'b-', linewidth = 2, label='y = 2x + 3')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('The function y = 2x + 3')
ax1.grid(True)
ax1.legend()

#Derivative plot
ax2.plot(x, dy_dx, 'r-',linewidth = 2, label='dy/dx = 2')
ax2.set_xlabel('x')
ax2.set_ylabel('dy/dx')
ax2.set_title('The derivative dy/dx = 2')
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()
```
